```python
import tensorflow as tf
import cv2


CATEGORIES = ["Dog","Cat"]

def prepare1(filepath):
    IMG_SIZE = 50
    img_array = cv2.imread(filepath,cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
    return new_array.reshape(-1,IMG_SIZE,IMG_SIZE,1)

def prepare2(filepath):
    IMG_SIZE = 100
    img_array = cv2.imread(filepath,cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
    return new_array.reshape(-1,IMG_SIZE,IMG_SIZE,1)

def prepare3(filepath):
    IMG_SIZE = 70
    img_array = cv2.imread(filepath,cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
    return new_array.reshape(-1,IMG_SIZE,IMG_SIZE,1)
model1 = tf.keras.models.load_model("64x3-CNN.keras")
model2 = tf.keras.models.load_model("64x3-CNN1.keras")
model3 = tf.keras.models.load_model("64x3-CNN2.keras")
```

```python
img_name = "social-logo.jpg"
prediction1 = model1.predict([prepare1(img_name)])
#print(prediction)
index1 = int(prediction1[0][0])
#print(index)
print("model1:"+CATEGORIES[index1])
prediction2 = model2.predict([prepare2(img_name)])
#print(prediction)
index2 = int(prediction2[0][0])
#print(index)
print("model2:"+CATEGORIES[index2])

prediction3 = model3.predict([prepare3(img_name)])
#print(prediction)
index3 = int(prediction3[0][0])
#print(index)
print("model3:"+CATEGORIES[index3])
```

```python
import os
DATADIR = r"C:\Users\KIIT\notebook\train\dogs"
miss = 0
for img in os.listdir(DATADIR):
    #print(os.path.join(DATADIR,img))
    pred = model3.predict([prepare3(os.path.join(DATADIR,img))])
    ans = CATEGORIES[int(pred[0][0])]
    #print(ans)
    if ans != "Dog":
        miss = miss + 1

print("miss:")
print(miss)
print("miss percentage: ")
print((miss/278)*100)
```
